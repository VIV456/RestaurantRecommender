```
import pandas as pd
import numpy as np
```

```
from surprise import Dataset, Reader
from surprise import SVD
from surprise import accuracy
from surprise.model_selection import cross_validate, train_test_split
from surprise.prediction_algorithms import Prediction
import random
from surprise.model_selection import GridSearchCV
```

```
df= pd.read_csv('toronto_reviews.csv')
```

```
del df['Unnamed: 0']
```

```
# Instantiate a 'Reader' to read in the data so Surprise can use it
reader = Reader(rating_scale=(1, 5))

# The columns must correspond to user id, item id and ratings (in that order).
data = Dataset.load_from_df(df[['user_id', 'business_id', 'stars_revs']], reader)
```

```
data.split(n_folds=5)
```

```
param_grid = {'n_epochs': [10,20,40], 'n_factors':[1,5], 'biased': [True, False], 'lr_all': [0.001, 0.005, 0.02], 'reg_all':[0.04,0.06,0.08], 'verbose': ['True']}
```

```
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=5)
```

```
gs.fit(data)
```

```
results = gs.cv_results
```

```
results_df = pd.DataFrame.from_dict(gs.cv_results)
results_df.to_csv('svd_results.csv')
```

```
#Of the top 4 results, all are 40 epochs, biased, lr= 0.005, reg is 0.08 or 0.06, n factors is 1 or 5
results_df.sort_values(by='mean_test_rmse').head()
```

```
param_grid_2 = {'n_epochs': [40, 80], 'n_factors':[1,3,5], 'biased': [True], 'lr_all': [0.005, 0.01], 'reg_all':[0.06,0.08,0.10], 'verbose': ['True']}
```

```
gs2 = GridSearchCV(SVD, param_grid_2, measures=['rmse', 'mae'], cv=5)
```

```
gs2.fit(data)
```

```
results2 = gs2.cv_results
```

```
results2_df = pd.DataFrame.from_dict(gs2.cv_results)
results2_df.to_csv('svd_refined_results.csv')
```

```
#best params
results2_df.sort_values(by='mean_test_rmse').head(1)
```
